# TRAPEZOIDAL DECOMPOSITION

The environment for incomplete BUG 2 algorithm

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(15,10))
# creating the rectangular arena
a=[0,5,5,0,0]
b=[0,0,5,5,0]
plt.plot(a, b, label="rectangular arena")

# now lets create the obstacles
obs1x=[1,4,4,1,1]
obs1y=[0.25,0.25,1.25,1.25,0.25]
plt.plot(obs1x, obs1y, label="obstacle-1")

obs2x=[2,5,5,2,2]
obs2y=[1.5,1.5,2.5,2.5,1.5]
plt.plot(obs2x, obs2y, label="obstacle-2")

obs3x=[0,4,4,0,0]
obs3y=[2.75,2.75,3.75,3.75,2.75]
plt.plot(obs3x, obs3y, label="obstacle-3")

obs4x=[2.75,3.15,1.15,0.75,2.75]
obs4y=[4.00,4.15,4.75,4.65,4.00]
plt.plot(obs4x, obs4y, label="obstacle-4")

# now let's do the decomposition
dec1x=[1,1]
dec1y=[0,2.75]
plt.plot(dec1x, dec1y, "--")

dec2x=[4,4]
dec2y=[0,1.5]
plt.plot(dec2x, dec2y, "--")

dec3x=[2,2]
dec3y=[1.25,2.75]
plt.plot(dec3x, dec3y, "--")

dec4x=[4,4]
dec4y=[2.5,5]
plt.plot(dec4x, dec4y, "--")

dec5x=[0.75,0.75]
dec5y=[3.75,5]
plt.plot(dec5x, dec5y, "--")

dec6x=[1.15,1.15]
dec6y=[4.75,5]
plt.plot(dec6x, dec6y, "--")

dec7x=[2.75,2.75]
dec7y=[3.75,4.00]
plt.plot(dec7x, dec7y, "--")

dec8x=[3.15,3.15]
dec8y=[3.75,5]
plt.plot(dec8x, dec8y, "--")

# reebx=[0.5, 2.5, 4.5, 3, 1.5, 3, 4.5, 3.575, 2.95, 2.15, 0.95, 0.375, 1.75, 2.95, 3.575, 4.5, 3, 1.5, 0.5]
# reeby=[1.375, 0.125, 0.75, 1.375, 2, 2.625, 3.75, 4.375, 3.9, 4.70, 4.87, 4.375, 4, 3.9, 4.375, 3.75, 2.625, 2, 1.375]
# plt.plot(reebx, reeby,"-o")

# plt.legend()
x_lst=np.arange(0,5.5,0.5)
y_lst=np.arange(0,5.5,0.25)
plt.xticks(x_lst)
plt.yticks(y_lst)
plt.title("Trapezoidal decomposition of the environment")
plt.grid()
plt.legend()

plt.show()
```
